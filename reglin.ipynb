1. Dataset

```python
X = [
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 8],
    [8, 8]
]

y = [0, 0, 0, 1, 1, 1]
```

```python
w1 = 0
w2 = 0
b = 0
lr = 0.01

prediction = []

for element in X:
    x1 = element[0]
    x2 = element[1]
    y_pred = w1 * x1 + w2 * x2 + b
    prediction.append(y_pred)
print(prediction)
```

```python
somme_erreur_carré = []
sme = []
n = len(X)

zip(y, prediction)
for element1, element2 in zip(y, prediction):
    y_i = element1
    y_pred_i = element2
    erreur_carré = (y_i - y_pred_i) ** 2
    somme_erreur_carré.append(erreur_carré)

sme = sum(somme_erreur_carré) / n
sme
```

decente de gradient 
$$\frac{\partial L}{\partial w_1} = \frac{2}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i) \cdot x_{1i}$$
$$\frac{\partial L}{\partial w_2} = \frac{2}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i) \cdot x_{2i}$$
$$\frac{\partial L}{\partial b} = \frac{2}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)$$

```python
somme_difference1 = []
liste_gradw1 = []
liste_gradw2 = []
liste_gradw3 = []

for element1, element2, element3 in zip(X, y, prediction):

    somme = element3 - element2
    somme_difference1.append(somme)

    gradw1 = (2/n * somme) * element1[0]
    liste_gradw1.append(gradw1)
    
    gradw2 = (2/n * somme) * element1[1]
    liste_gradw2.append(gradw2)

    gradw3 = (2/n * somme)
    liste_gradw3.append(gradw3)


grad_w1 = sum(liste_gradw1) / n
grad_w2 = sum(liste_gradw2) / n
grad_b  = sum(liste_gradw3) / n

print(grad_w1, grad_w2, grad_b)
```

```python
w1 = w1 - lr * grad_w1
w2 = w2 - lr * grad_w2
b  = b  - lr * grad_b

print(w1,w2,b)
```

```python
lr = 0.01
w1 = 0
w2 = 0
b = 0
n = len(X)

histo_loss = []

for i in range(1000):
    
    prediction = []
    for element in X:
        x1 = element[0]
        x2 = element[1]
        y_pred = w1 * x1 + w2 * x2 + b
        prediction.append(y_pred)
    somme_erreur_carré = []
    for element1, element2 in zip(y, prediction):
        erreur_carré = (element1 - element2) ** 2
        somme_erreur_carré.append(erreur_carré)
    sme = sum(somme_erreur_carré) / n
    histo_loss.append(sme)    

    liste_gradw1 = []
    liste_gradw2 = []
    liste_gradw3 = []
    for element1, element2, element3 in zip(X, y, prediction):
        somme = element3 - element2
        liste_gradw1.append((2/n * somme) * element1[0])
        liste_gradw2.append((2/n * somme) * element1[1])
        liste_gradw3.append((2/n * somme))

    grad_w1 = sum(liste_gradw1)
    grad_w2 = sum(liste_gradw2)
    grad_b  = sum(liste_gradw3)

    w1 = w1 - lr * grad_w1
    w2 = w2 - lr * grad_w2
    b  = b  - lr * grad_b

print(i, sme)
sns.lineplot(x=range(1000), y=histo_loss)
```

```python
correct = 0
for pred, vrai in zip(prediction, y):
    if pred >= 0.5:
        pred_arrondi = 1
    else:
        pred_arrondi = 0
    if pred_arrondi == vrai:
        correct += 1

accuracy = correct / n * 100
print(accuracy)
```
